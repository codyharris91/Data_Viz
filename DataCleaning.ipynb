```python
def adjust_cols(df):
    #merge device and drug columns - then drop them
    for i in range(1,6):
        first = f'Name_of_Associated_Covered_Drug_or_Biological{i}'
        second = f'Name_of_Associated_Covered_Device_or_Medical_Supply{i}'
        col_name = f'Name_of_Drug_or_Biological_or_Device_or_Medical_Supply_{i}'
        df[col_name] = df[first] + '--' + df[second]
        df = df.drop(columns = [first, second])

    #Rename columns to know 
    re_col = {'Product_Indicator': 'Related_Product_Indicator',
              'NDC_of_Associated_Covered_Drug_or_Biological1': 'Associated_Drug_or_Biological_NDC_1',
              'NDC_of_Associated_Covered_Drug_or_Biological2': 'Associated_Drug_or_Biological_NDC_2',
              'NDC_of_Associated_Covered_Drug_or_Biological3': 'Associated_Drug_or_Biological_NDC_3',
              'NDC_of_Associated_Covered_Drug_or_Biological4': 'Associated_Drug_or_Biological_NDC_4',
              'NDC_of_Associated_Covered_Drug_or_Biological5': 'Associated_Drug_or_Biological_NDC_5'}
    df = df.rename(columns = re_col)
    return df
```

## Aggregate by Payment Amount

```python
def agg_by_payment(group_col, col_names):
    years = ['2013', '2014', '2015', '2016', '2017', '2018', '2019']
    #states = [None] * len(years)
    agg_op_features = {'Total_Amount_of_Payment_USDollars': ['sum', 'count', 'mean']}
    agg_df = pd.DataFrame()
    for i, year in enumerate(years):
        if 'NA' not in datasets[year][group_col].cat.categories:
            datasets[year][group_col] = datasets[year][group_col].cat.add_categories('NA').fillna('NA')
        st = datasets[year].groupby(group_col).agg(agg_op_features)
        st['Year'] = int(year)
        st.columns = st.columns.map('_'.join)
        st = st.reset_index()
        if(agg_df.empty):
            agg_df = st
        else:
            agg_df = agg_df.append(st)
    agg_df['Total_Amount_of_Payment_USDollars_mean'] = np.round(agg_df['Total_Amount_of_Payment_USDollars_mean'], 2)
    agg_df['Total_Amount_of_Payment_USDollars_sum'] = np.round(agg_df['Total_Amount_of_Payment_USDollars_sum'], 2)
    agg_df.columns = col_names
    agg_df.head(15)
    return agg_df
```

## Aggregated by Drug

```python
# count the number of non-nulls
years = ['2013', '2014', '2015', '2016', '2017', '2018', '2019']
drug_cols =['Name_of_Drug_or_Biological_or_Device_or_Medical_Supply_1', 'Name_of_Drug_or_Biological_or_Device_or_Medical_Supply_2',
            'Name_of_Drug_or_Biological_or_Device_or_Medical_Supply_3', 'Name_of_Drug_or_Biological_or_Device_or_Medical_Supply_4',
            'Name_of_Drug_or_Biological_or_Device_or_Medical_Supply_5']
agg_op_features = {'Total_Amount_of_Payment_USDollars': ['sum', 'count', 'mean'],
                   'amount_split': ['sum', 'count', 'mean']}
long = 'Name_of_Drug_or_Biological_or_Device_or_Medical_Supply_1'
drug_agg_df = pd.DataFrame()

for year in years:
    if 'drug_count' not in datasets[year]:
        datasets[year]['drug_count'] = datasets[year][drug_cols].notna().sum(axis=1)
    if 'NONE' not in datasets[year][long].cat.categories:
        datasets[year][long] = datasets[year][long].cat.add_categories('NONE')
    datasets[year][long][datasets[year]['drug_count'] == 0] = 'NONE'
    datasets[year]['drug_count'] = datasets[year]['drug_count'].replace(0, 1)
    drug_melt = pd.melt(datasets[year], id_vars = ['Total_Amount_of_Payment_USDollars', 'drug_count', 'Program_Year'], value_vars = drug_cols,
                        var_name = 'Slot', value_name = 'Drug')
    drug_melt['amount_split'] = drug_melt['Total_Amount_of_Payment_USDollars'] / drug_melt['drug_count']
    lel = drug_melt.groupby('Drug', as_index=False).agg(agg_op_features)
    lel.columns = lel.columns.map('_'.join)
    lel['Year'] = year
    if(drug_agg_df.empty):
        drug_agg_df = lel
    else:
        drug_agg_df = drug_agg_df.append(lel)

drug_agg_df.to_csv(f'Data/Output/drug_agg.csv', index = False)
```

```python
years = ['2013', '2014', '2015', '2016', '2017', '2018', '2019']
for year in years:
    trav = datasets[year][cols].loc[datasets[year]['City_of_Travel'].notna() |
                                    datasets[year]['State_of_Travel'].notna() |
                                    datasets[year]['Country_of_Travel'].notna()]
    #Drop any row where the country is US, but state is null. Can't really know what the real state should be in many cases
    #Not dropping many rows at all
    drop_ind = trav[(trav['State_of_Travel'].isnull()) & (trav['Country_of_Travel'] == 'United States')].index
    trav.drop(drop_ind, inplace = True)
    drop_ind2 = trav[(trav['Recipient_State'].isnull()) & (trav['Recipient_Country'] == 'United States')].index
    trav.drop(drop_ind2, inplace = True)
    drop_ind3 = trav[(trav['Recipient_City'].isnull()) & ((trav['Recipient_Country'] != 'United States') | (trav['Recipient_Country'].isnull()))].index
    trav.drop(drop_ind3, inplace = True)
    drop_ind4 = trav[(trav['City_of_Travel'].isnull()) & ((trav['Country_of_Travel'] != 'United States') | (trav['Country_of_Travel'].isnull()))].index
    trav.drop(drop_ind4, inplace = True)
    trav['Recipient_City'] = trav['Recipient_City'].astype(str)
    trav['Recipient_City'] = trav['Recipient_City'].str.upper()
    trav['City_of_Travel'] = trav['City_of_Travel'].astype(str)
    trav['City_of_Travel'] = trav['City_of_Travel'].str.upper()
    trav['Recipient_Country'] = trav['Recipient_Country'].astype(str)
    trav['Recipient_Country'] = trav['Recipient_Country'].str.upper()
    trav['Country_of_Travel'] = trav['Country_of_Travel'].astype(str)
    trav['Country_of_Travel'] = trav['Country_of_Travel'].str.upper()
    trav['Recipient_State'] = trav['Recipient_State'].astype(str)
    trav['State_of_Travel'] = trav['State_of_Travel'].astype(str)
    trav['Source_name'] = ''
    trav['Target_name'] = ''
    trav['Source_name'][trav['Recipient_Country'] == 'UNITED STATES'] = trav['Recipient_City'] + ', ' + trav['Recipient_State']
    trav['Source_name'][trav['Recipient_Country'] != 'UNITED STATES'] = trav['Recipient_City'] + ', ' + trav['Recipient_Country']
    trav['Target_name'][trav['Country_of_Travel'] == 'UNITED STATES'] = trav['City_of_Travel'] + ', ' + trav['State_of_Travel']
    trav['Target_name'][trav['Country_of_Travel'] != 'UNITED STATES'] = trav['City_of_Travel'] + ', ' + trav['Country_of_Travel']
    unique_places = pd.unique(trav[['Source_name', 'Target_name']].values.ravel('K'))
    up_df = pd.DataFrame({'Label': unique_places})
    up_df.index.name = 'Id'
    up_df.to_csv(f'Data/Output/Network/node_{year}.csv')
    up_df.reset_index(level=0, inplace=True)
    trav = trav.merge(up_df, left_on = 'Source_name', right_on = 'Label', how = 'left')
    trav.rename(columns = {'Id': 'Source'}, inplace = True)
    trav = trav.merge(up_df, left_on = 'Target_name', right_on = 'Label', how = 'left')
    source = trav['Source']
    target = trav['Id']
    edge = pd.DataFrame({'Source': source, 'Target': target})
    edge_df = edge.groupby(['Source', 'Target']).size().reset_index(name='Weight')
    same_city = edge_df[edge_df['Source'] == edge_df['Target']]
    same_city.to_csv(f'Data/Output/Network/Extra/same_city_{year}.csv', index = False)
    edge_df = edge_df[edge_df['Source'] != edge_df['Target']]
    edge_df['Type'] = 'Undirected'
    edge_df.to_csv(f'Data/Output/Network/edge_{year}.csv', index = False)
```
