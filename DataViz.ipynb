## Counts

```python
h = 200
w = 324
v_dis = ['#482575', '#414487', '#35608d', '#2a788e', '#21918d', '#22a884', '#43bf71', '#7ad151', '#bcdf27']

t = alt.Chart(count_url).mark_line(point=True).encode(
    x='Year:O',
    y=alt.Y('Total:Q', scale = alt.Scale(domain = (3500000, 12000000))),
    tooltip = ['Year:O', alt.Tooltip('Total:Q', title = 'Total Records', format = ',.0f')],
    color = alt.value(v_dis[0])
).properties(
    width=w,
    height=h,
    title='Total Number of Records'
)

tp = alt.Chart(count_url).mark_line(point=True).encode(
    x=alt.X('Year:O'),
    y=alt.Y('Total_Paid:Q', title = 'USD', scale = alt.Scale(domain = (500000000, 4000000000)), axis=alt.Axis(format='$,f')),
    tooltip = ['Year:O', alt.Tooltip('Total_Paid:Q', title = 'Amount Paid', format = '$,.2f')],
    color = alt.value(v_dis[4])
).properties(
    width=w,
    height=h,
    title = 'Total Amount Paid'
)

doc = alt.Chart(count_url).mark_line(point=True).encode(
    x='Year:O',
    y=alt.Y('Doctors:Q', scale = alt.Scale(domain = (400000, 700000))),
    tooltip = ['Year:O', alt.Tooltip('Doctors:Q', title = 'Number of Doctors', format = ',.0f')],
    color = alt.value(v_dis[2])

).properties(
    width=w,
    height=h,
    title = 'Number of Doctors Paid'
)

cc = alt.Chart(count_url).mark_line(point=True).encode(
    x='Year:O',
    y=alt.Y('Company_Count:Q', title = 'Number of Companies', scale = alt.Scale(domain = (1300, 1700))),
    tooltip = ['Year:O', alt.Tooltip('Company_Count:Q', title = 'Number of Companies', format = ',.0f')],
    color = alt.value(v_dis[6])
).properties(
    width=w,
    height=h,
    title = 'Companies that Made Payments to Doctors'
)

col1 = alt.vconcat(t, tp)
col2 = alt.vconcat(doc, cc)
all_c = alt.hconcat(col1, col2).configure_point(size=80)
all_c
```

```python
all_c.save('data_counts.html')
```

## Nature and Form Heatmap

```python
w = 100
h = 400
years = ['2014', '2015', '2016', '2017', '2018', '2019']
charts = [None] * (len(years) + 1)
charts[0] = alt.Chart(heat_url, title = '2013').mark_rect().encode(
    x=alt.X('Form:O', axis = alt.Axis(title = ' ')),
    y=alt.Y('Nature:O', axis = alt.Axis(title = ' ')),
    color = alt.Color('Count:Q',scale=alt.Scale(type='log',scheme='viridis')),
    tooltip = ['Form:O', 'Nature:O', alt.Tooltip('Count:Q', title = 'Count', format = ',')],
).properties(
    width = w,
    height = h
).transform_filter(
    datum['Year'] == '2013'
)
for i, year in enumerate(years):
    if(year in ['2014', '2015']):
        w = 100
        h = 400
    else:
        w = 150
        h = 400
    charts[i+1] = alt.Chart(heat_url, title = year).mark_rect().encode(
        x=alt.X('Form:O', axis = alt.Axis(title = ' ')),
        y=alt.Y('Nature:O', axis = alt.Axis(labels = False, title = ' ')),
        color = alt.Color('Count:Q',scale=alt.Scale(type='log',scheme='viridis')),
        tooltip = ['Form:O', 'Nature:O', alt.Tooltip('Count:Q', title = 'Count', format = ',')],
    ).properties(
        width = w,
        height = h
    ).transform_filter(
        datum['Year'] == year
    )
heat_map_chart = alt.hconcat(charts[0], charts[1], charts[2], charts[3], charts[4], charts[5], charts[6]).configure_axis(labelLimit = 0, labelFontSize = 11)
heat_map_chart
```
